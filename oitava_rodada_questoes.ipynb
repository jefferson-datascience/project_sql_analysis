**Exercício:** Montar a grande tabela da Olist, a partir da união de todas as tabelas.

**Observação:** Foi desconsiderado da grande tabela a tabela geolocation. Ela gera a duplicidade de todos os orders id de forma desnecessária. Assim, ela foi desconsiderada. 

Provavelmente, deve haver alguma outra key que permita essa união. Mas com o que foi fornecido pela plataforma kaggle, não há possibilidade de considerar a tabela geolocation na tabela grande.

```python
query = """ 
    SELECT
        *
    FROM 
        order_reviews or2 LEFT JOIN  orders o          ON (or2.order_id = o.order_id)
                          INNER JOIN order_items oi    ON (o.order_id = oi.order_id) 
                          LEFT JOIN  sellers s         ON (oi.seller_id  = s.seller_id) 
                          INNER JOIN customer c        ON (o.customer_id = c.customer_id)
                          LEFT JOIN order_payments op  ON (o.order_id = op.order_id)
                          INNER JOIN products p        ON (oi.product_id = p.product_id)
    LIMIT 100
    """

df_question = pd.read_sql(query, cursor)
df_question.to_csv('files/diretorio_oitava_rodada_questoes/question.csv')

df_question
```
