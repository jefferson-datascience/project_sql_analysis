## 1ª Rodada de Questões do CEO

### Bibliotecas

```python
import pandas as pd
from sqlalchemy import create_engine
```

### Acessando o Banco de Dados

```python
# Conexão com o Banco de Dados
engine = create_engine('sqlite:///db_olist.sqlite')

# Cursor para realizar as querys
cursor = engine.connect()
```

**1.** Gere uma tabela com o id do cliente, a cidade e o estado onde ele vive.

```python
query = """
    SELECT 
        c.customer_id, c.customer_city, c.customer_state
    FROM 
        customer c 
    LIMIT 
        10
        """

# Salvando o arquivo
df_question_1 = pd.read_sql(query, cursor)
df_question_1.to_csv('files/diretorio_primeira_rodada_questoes/question_1.csv')

df_question_1
```

**2.** Gere uma tabela com o id do cliente e a cidade, somente dos cliente que vivem em Santa Catarina.

```python
query = """
    SELECT 
        c.customer_id, c.customer_city 
    FROM
        customer c 
    WHERE 
        c.customer_state = 'SC' 
    LIMIT
        10
    """

# Salvando o arquivo
df_question_2 = pd.read_sql(query, cursor)
df_question_2.to_csv('files/diretorio_primeira_rodada_questoes/question_2.csv')

df_question_2
```

**3.** Gere somente um tabela com o id do cliente e o estado, somente dos clientes que vivem em Florianópolis.

```python
query = """
    SELECT 
        c.customer_id, c.customer_city, c.customer_state 
    FROM 
        customer c 
    WHERE
        c.customer_city = 'florianopolis'
    LIMIT 10
"""

# Salvando o arquivo
df_question_3 = pd.read_sql(query, cursor)
df_question_3.to_csv('files/diretorio_primeira_rodada_questoes/question_3.csv')

df_question_3
```

**4.** Gere uma tabela com o estado, latitude e longitude do estado de São Paulo.

```python
query = """
    SELECT
        g.geolocation_state, g.geolocation_lat, g.geolocation_lng 
    FROM
        geolocation g 
    WHERE 
        g.geolocation_state = 'SP'
    LIMIT 
        10
"""

# Salvando o arquivo
df_question_4 = pd.read_sql(query, cursor)
df_question_4.to_csv('files/diretorio_primeira_rodada_questoes/question_4.csv')

df_question_4
```

**5.** Gere uma tabela com o id do produto, a datade envio e o preço, somente para produtos acima de 6300.

```python
query = """
    SELECT
        oi.product_id, oi.shipping_limit_date, oi.price  
    FROM
        order_items oi  
    WHERE 
        oi.price > 6300
"""

# Salvando o arquivo
df_question_5 = pd.read_sql(query, cursor)
df_question_5.to_csv('files/diretorio_primeira_rodada_questoes/question_5.csv')

df_question_5
```

**6.** Gere uma tabela com o id do pedido, o tipo de pagamento e o número de parcelas, somente para produtos com parcelas menores que 1.

```python
query = """
    SELECT
        op.order_id, op.payment_type, op.payment_installments 
    FROM 
        order_payments op 
    WHERE
        op.payment_installments < 1
"""

# Salvando o arquivo
df_question_6 = pd.read_sql(query, cursor)
df_question_6.to_csv('files/diretorio_primeira_rodada_questoes/question_6.csv')

df_question_6
```

**7.** Gere uma tabela com o id do pedido, id do cliente, o status do pedido e a data de aprovação, somente para compras aprovadas até o dia 05 de Maio de 2016.

```python
query = """
    SELECT
        o.order_id, o.customer_id, o.order_status, o.order_approved_at 
    FROM 
        orders o 
    WHERE 
        o.order_approved_at <= '2016-10-05'
"""

# Salvando o arquivo
df_question_7 = pd.read_sql(query, cursor)
df_question_7.to_csv('files/diretorio_primeira_rodada_questoes/question_7.csv')

df_question_7
```
