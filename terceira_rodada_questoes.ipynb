### Acessando o Banco de Dados

```python
# Conexão com o Banco
engine = create_engine('sqlite:///db_olist.sqlite')

# Criando a conexão com o Banco
cursor = engine.connect()
```

**1.** Qual o número de clientes únicos de todos os estados?

```python
query = """
    SELECT
        c.customer_state , COUNT(DISTINCT c.customer_unique_id) AS totat_customer_by_state 
    FROM 
        customer c 
    GROUP BY
        c.customer_state 
        """

# Salvando o arquivo
df_question_1 = pd.read_sql(query, cursor)
df_question_1.to_csv('files/diretorio_terceira_rodada_questoes/question_1.csv')

df_question_1
```

**2.** Qual o número de cidades únicas de todos os estados?

```python
query = """
    SELECT
         g.geolocation_state, COUNT( DISTINCT g.geolocation_city ) AS num_unique_city_by_state
    FROM
        geolocation g 
    GROUP BY
        g.geolocation_state 
        """

# Salvando o arquivo
df_question_2 = pd.read_sql(query, cursor)
df_question_2.to_csv('files/diretorio_terceira_rodada_questoes/question_2.csv')

df_question_2
```

**3.** Qual o número de clientes únicos por estado e por cidade?

```python
query = """
    SELECT
         c.customer_state, c.customer_city, COUNT(DISTINCT c.customer_id) AS num_unique_customer_city_and_state  
    FROM 
        customer c 
    GROUP BY
        c.customer_state, c.customer_city 
    LIMIT 50
        """

# Salvando o arquivo
df_question_3 = pd.read_sql(query, cursor)
df_question_3.to_csv('files/diretorio_terceira_rodada_questoes/question_3.csv')

df_question_3
```

**5.** Qual o número total de pedidos únicos acima de 3.500 por cada vendedor?

```python
query = """
    SELECT 
        oi.seller_id, oi.order_id, COUNT(DISTINCT oi.order_id) AS num_order_by_seller
    FROM 
        order_items oi 
    WHERE
        oi.price > 3500
    GROUP BY 
        oi.seller_id 
        """

pd.read_sql(query, cursor)# Salvando o arquivo
df_question_5 = pd.read_sql(query, cursor)
df_question_5.to_csv('files/diretorio_terceira_rodada_questoes/question_5.csv')

df_question_5
```

**6.** Qual o número total de pedidos únicos, a data mínima e máxima de envio, o valor máximo, mínimo e médio do frete dos pedidos acima de R$ 1.100 por cada vendedor?

```python
query = """
    SELECT 
        oi.seller_id, 
        COUNT(DISTINCT oi.order_id) AS num_orders,
        MIN(oi.shipping_limit_date) AS min_shipping_limit_date,
        MAX(oi.shipping_limit_date) AS max_shipping_limit_date,
        MIN(oi.freight_value) AS max_freight_value,
        MAX(oi.freight_value) AS min_freight_value,
        AVG(oi.freight_value) AS avg_freight_value 
    FROM 
        order_items oi 
    WHERE
        oi.price > 1100
    GROUP BY
        oi.seller_id 
    LIMIT 50
        """

# Salvando o arquivo
df_question_6 = pd.read_sql(query, cursor)
df_question_6.to_csv('files/diretorio_terceira_rodada_questoes/question_6.csv')

df_question_6
```

**8.** Qual a quantidade de vendedores distintos que receberam algum pedido antes do dia 23 de setembro de 2016 e qual foi o preço médio desses pedidos?

```python
query = """
    SELECT
        COUNT(DISTINCT oi.seller_id) AS num_sellers,
        AVG(oi.price) AS avg_price 
    FROM 
        order_items oi 
    WHERE
        oi.shipping_limit_date < '2016-09-23'
    GROUP BY
        oi.shipping_limit_date 
        """

# Salvando o arquivo
df_question_8 = pd.read_sql(query, cursor)
df_question_8.to_csv('files/diretorio_terceira_rodada_questoes/question_8.csv')

df_question_8
```

**9.** Qual a quantidade de pedidos por tipo de pagamentos?

**10.** Qual a quantidade de pedidos, a média do valor do pagamento e o número máximo de parcelas por tipo de pagamentos?

```python
query = """
    SELECT 
        op.payment_type, 
        COUNT(op.order_id)           AS total_order,
        AVG(op.payment_value)        AS mean_value_payment,
        MAX(op.payment_installments) AS max_payments_installments 
    FROM 
        order_payments op  
    GROUP BY
        op.payment_type  
        """

# Salvando o arquivo
df_question_10 = pd.read_sql(query, cursor)
df_question_10.to_csv('files/diretorio_terceira_rodada_questoes/question_10.csv')

df_question_10
```

**11.** Qual a valor mínimo, máximo, médio e as soma total paga por cada tipo de pagamento e número de parcelas disponíveis?

```python
query = """
    SELECT
        op.payment_type, op.payment_installments, 
        MIN(op.payment_value)        AS min_payments_value,
        MAX(op.payment_value)        AS max_payments_value,
        AVG(op.payment_value)        AS avg_payments_value,
        SUM(op.payment_value)        AS sum_payments_values 
    FROM 
        order_payments op 
    GROUP BY
        op.payment_type, op.payment_installments
    
        """

# Salvando o arquivo
df_question_11 = pd.read_sql(query, cursor)
df_question_11.to_csv('files/diretorio_terceira_rodada_questoes/question_11.csv')

df_question_11
```

**12.** Qual a média de pedidos por cliente?

```python
query = """
    SELECT 
        tabela.customer_id, AVG(tabela.num_orders) AS average_orders
    FROM 
        (SELECT
            o.customer_id , COUNT(o.order_id) AS num_orders
        FROM 
            orders o
        GROUP BY 
         o.customer_id ) AS tabela
    GROUP BY 
        tabela.customer_id
    LIMIT 50
        """

# Salvando o arquivo
df_question_12 = pd.read_sql(query, cursor)
df_question_12.to_csv('files/diretorio_terceira_rodada_questoes/question_12.csv')

df_question_12
```

**13.** Qual a quantidade de pedidos realizados por cada status do pedido, a partir do dia 23 de Setembro de 2016?

```python
query = """
        SELECT
        o.order_status,
        COUNT(o.order_id) AS num_orders 
    FROM
        orders o 
    WHERE
        o.order_approved_at > '2016-09-23 00:00:00'
    GROUP BY 
        o.order_status
        """

# Salvando o arquivo
df_question_13 = pd.read_sql(query, cursor)
df_question_13.to_csv('files/diretorio_terceira_rodada_questoes/question_13.csv')

df_question_13
```

**14.** Qual a quantidade de pedidos realizados por dia, a partir do dia 23 de Setembro de 2016?

```python
query = """
    SELECT 
        DATE(o.order_approved_at) AS date_,
        COUNT(o.order_id) AS total_orders_approved
    FROM
        orders o 
    WHERE 
        o.order_approved_at > '2016-09-23 00:00:00' 
    GROUP BY
        DATE(o.order_approved_at) 
    LIMIT 50
        """

# Salvando o arquivo
df_question_14 = pd.read_sql(query, cursor)
df_question_14.to_csv('files/diretorio_terceira_rodada_questoes/question_14.csv')

df_question_14
```
