### Bibliotecas

```python
from sqlalchemy import create_engine
import pandas as pd
```

### Acessando o Banco de Dados

```python
# Conexão com o Banco
engine = create_engine('sqlite:///db_olist.sqlite')

# Criando a conexão com o Banco
cursor = engine.connect()
```

**1.** Gerar uma tabela de dados com 10 linhas, contendo o id do pedido, o id do cliente, o status do pedido, o id do produto e o preço do produto.

```python
query = """
    SELECT 
        o.order_id, o.customer_id, o.order_status, oi.product_id, oi.price  
    FROM 
        orders o INNER JOIN order_items oi ON (o.order_id = oi.order_id)
    LIMIT 
        10
        """

# Salvando o arquivo
df_question_1 = pd.read_sql(query, cursor)
df_question_1.to_csv('files/diretorio_setima_rodada_questoes/question_1.csv')

df_question_1
```

**2.** Gerar uma tabela de dados com 20 linhas, contendo o id do pedido, o estado do cliente, a cidade do cliente, o status do pedido, o id do produto e o preço do produto, somente para clientes do estado de São Paulo.


```python
query = """
    SELECT 
        o.order_id, c.customer_state, c.customer_city, o.order_status, oi.product_id, oi.price  
    FROM
        order_items oi INNER JOIN orders o   ON (oi.order_id = o.order_id)
                       INNER JOIN customer c ON (o.customer_id = c.customer_id)	
    WHERE 
        c.customer_state = 'SP'
    LIMIT 
        20
        """

df_question_2 = pd.read_sql(query, cursor)
df_question_2.to_csv('files/diretorio_setima_rodada_questoes/question_2.csv')

df_question_2
```

**3.** Gerar uma tabela de dados com 50 linhas, contendo o id do pedido, o estado e a cidade do cliente, o status do pedido, o nome da categoria do produto e o preço do produto, somente para pedido com o status igual a cancelado.


```python
query = """
        SELECT 
        o.order_id, c.customer_state, c.customer_city, o.order_status, p.product_category_name, oi.price  
    FROM 
        products p INNER JOIN order_items oi ON (p.product_id = oi.product_id)
                   INNER JOIN orders o       ON (oi.order_id = o.order_id)
                   INNER JOIN customer c     ON (o.customer_id = c.customer_id)
    WHERE 
        o.order_status = 'canceled'
    LIMIT 
        50
        """

df_question_3 = pd.read_sql(query, cursor)
df_question_3.to_csv('files/diretorio_setima_rodada_questoes/question_3.csv')

df_question_3
```

**4.** Gerar uma tabela de dados com 80 linhas, contendo o id do pedido, o estado e a cidade do cliente, o status do pedido, o nome da categoria do produto, o preço do produto, a cidade e o estado do vendedor e a data de aprovação do pedido, somente para os pedido aprovados a patir do dia 16 de setembro de 2016


```python
query = """
         SELECT 
        o.order_id, c.customer_state, c.customer_city, o.order_status, p.product_category_name, oi.price, g.geolocation_city, g.geolocation_state, o.order_approved_at  
    FROM 
        products p INNER JOIN order_items oi ON (p.product_id = oi.product_id)
                   INNER JOIN orders o       ON (oi.order_id = o.order_id)
                   INNER JOIN customer c     ON (o.customer_id = c.customer_id)
                   INNER JOIN geolocation g  ON (c.customer_zip_code_prefix = g.geolocation_zip_code_prefix)
    WHERE 
        DATE(o.order_approved_at) > '2016-09-16' 
    LIMIT 
        80
        """

df_question_4 = pd.read_sql(query, cursor)
df_question_4.to_csv('files/diretorio_setima_rodada_questoes/question_4.csv')

df_question_4
```

**5.** Gerar uma tabela de dados com 10 linhas, contendo o id do pedido, o estado e a cidade do cliente, o status do pedido, o nome da categoria do produto, o preço do produto, a cidade e o estado do vendedor, a data de aprovação do pedido e o tipo de pagamento, somente para o tipo de pagamento igual a boleto.


```python
query = """
    SELECT 
        o.order_id, c.customer_state, c.customer_city,  o.order_status, p.product_category_name, oi.price, g.geolocation_city, g.geolocation_state, o.order_approved_at, op.payment_type  
    FROM 
        geolocation g INNER JOIN customer c        ON (g.geolocation_zip_code_prefix = c.customer_zip_code_prefix)
                      INNER JOIN orders o          ON (c.customer_id = o.customer_id)
                      INNER JOIN order_payments op ON (o.order_id = op.order_id)
                      INNER JOIN order_items oi    ON (o.order_id = oi.order_id)
                      INNER JOIN products p        ON (oi.product_id = p.product_id)
    WHERE 
        op.payment_type = 'boleto'
    LIMIT 
        10
        """

df_question_5 = pd.read_sql(query, cursor)
df_question_5.to_csv('files/diretorio_setima_rodada_questoes/question_5.csv')

df_question_5
```


**6.** Gerar uma tabela de dados com 70 linhas, contendo o id do pedido, o estado e a cidade do cliente, o status do pedido, o nome da categoria do produto, o preço do produto, a cidade e o estado vendedor, a data de aprovação do pedido, tipo de pagamento, note de avaliação do produto, somente para os pedidos com a avaliação do produto igual a 1.

```python
query = """
       SELECT
        o.order_id, c.customer_state , c.customer_city , o.order_status , p.product_category_name , oi.price , s.seller_city , s.seller_state , o.order_approved_at , op.payment_type , or2.review_score 
    FROM 
        orders o INNER JOIN order_reviews or2 ON (o.order_id = or2.order_id)
                 INNER JOIN order_payments op ON (o.order_id = op.order_id)
                 INNER JOIN customer c        ON (o.customer_id = c.customer_id)
                 INNER JOIN order_items oi    ON (o.order_id = oi.order_id)
                 INNER JOIN products p        ON (oi.product_id = p.product_id)
                 INNER JOIN sellers s         ON (oi.seller_id = s.seller_id)
    WHERE 
        or2.review_score = 1
    LIMIT 70
        """

df_question_6 = pd.read_sql(query, cursor)
df_question_6.to_csv('files/diretorio_setima_rodada_questoes/question_6.csv')

df_question_6
```
