## 4ª Rodada de Questões do CEO

### Bibliotecas

```python
from sqlalchemy import create_engine
import pandas as pd
```

### Acessando o Banco de Dados

```python
# Conexão com o Banco
engine = create_engine('sqlite:///db_olist.sqlite')

# Criando a conexão com o Banco
cursor = engine.connect()
```

**1.** Qual o número de clientes únicos do estado de São Paulo?

```python
query = """
    SELECT
        COUNT(DISTINCT c.customer_id) AS num_unique_customer_sp 
    FROM
        customer c 
    WHERE
        c.customer_state = 'SP'
        """

# Salvando o arquivo
df_question_1 = pd.read_sql(query, cursor)
df_question_1.to_csv('files/diretorio_quarta_rodada_questoes/question_1.csv')

df_question_1
```

**2.** Qual o número total de pedidos únicos feitos no dia 08 de Outubro de 2016.

```python
query = """
    SELECT
        COUNT(DISTINCT oi.order_id) AS num_unique_orders 
    FROM 
        order_items oi 
    WHERE
        DATE(oi.shipping_limit_date) = '2016-10-08' 
        """

# Salvando o arquivo
df_question_2 = pd.read_sql(query, cursor)
df_question_2.to_csv('files/diretorio_quarta_rodada_questoes/question_2.csv')

df_question_2
```

**3.** Qual o número total de pedidos únicos feitos a partir do dia 08 de Outubro de 2016 .

```python
query = """
    SELECT
        COUNT(DISTINCT oi.order_id) AS num_unique_orders 
    FROM
        order_items oi 
    WHERE
        DATE(oi.shipping_limit_date) > '2016-10-08' 
        """

# Salvando o arquivo
df_question_3 = pd.read_sql(query, cursor)
df_question_3.to_csv('files/diretorio_quarta_rodada_questoes/question_3.csv')

df_question_3
```

**4.** Qual o número total de pedidos únicos feitos a partir do dia 08 de Outubro de 2016 incluso.

```python
query = """
    SELECT
        COUNT(DISTINCT oi.order_id) AS num_unique_orders 
    FROM
        order_items oi 
    WHERE 
        DATE(oi.shipping_limit_date) >= '2016-10-08' 
        """

# Salvando o arquivo
df_question_4 = pd.read_sql(query, cursor)
df_question_4.to_csv('files/diretorio_quarta_rodada_questoes/question_4.csv')

df_question_4
```

**5.** Qual o número total de pedidos únicos, a data mínima e máxima de envio, o valor máximo, mínimo e médio do frete dos pedidos abaixo de 1.100 por cada vendedor?


```python
query = """
    SELECT
        oi.seller_id, 
        COUNT(DISTINCT oi.order_id)       AS num_unique_orders, 
        MIN(DATE(oi.shipping_limit_date)) AS min_shipping_limit_date,
        MAX(DATE(oi.shipping_limit_date)) AS max_shipping_limit_date,
        MIN(oi.freight_value)             AS min_freight_value,
        MAX(oi.freight_value)             AS max_freight_value,
        AVG(oi.freight_value)             AS avg_freight_value
    FROM 
        order_items oi 
    WHERE
        oi.price < 1100
    GROUP BY 
        oi.seller_id
    LIMIT 50
        """

# Salvando o arquivo
df_question_5 = pd.read_sql(query, cursor)
df_question_5.to_csv('files/diretorio_quarta_rodada_questoes/question_5.csv')

df_question_5
```

**6.** Qual o número total de pedidos únicos, a data mínima e máxima de envio, o valor máximo, mínimo e médio do frete dos pedidos abaixo de 1.100 incluso por cada vendedor
